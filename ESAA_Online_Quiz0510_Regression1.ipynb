# **1. 단순 선형 회귀 분석**
- 전복의 나이를 예측하는 선형회귀모델을 생성하세요.
- 전복의 ‘성별’, ‘키’, ‘지름’, ‘높이’, ‘전체무게’, ‘몸통무게’, ‘내장무게’, ‘껍질무게’를 이용해 ‘껍질의 고리 수’를 예측한 뒤, **예측된 ‘껍질의 고리 수’에 1.5를 더하면 전복의 나이**가 됩니다.

```python
#구글 코랩 환경 사용으로 드라이브 마운트 및 디렉터리 설정
from google.colab import drive
drive.mount('/content/gdrive/')

path = "/content/gdrive/My Drive/ESAAData/"
```

```python
# 기본 모듈 불러오기
import numpy as np
import pandas as pd
```

**1) 데이터 load 및 변형**

```python
# 데이터 로드
data = pd.read_csv(path + "abalone.csv")
data.head()
print(data.shape)

# 성별 M은 Male, F는 Female, I는 Infant 이므로 따로 열 만들기
for label in "MFI":
    data[label] = data["Sex"] == label
data.drop('Sex', axis=1, inplace=True)
```

```python
data.head()
```

**2) X, y 선택**
: y는 Rings열, X는 Rings열을 제외한 나머지를 선택하되 전부 실수가 되도록 한다.

```python
# X,y 데이터 선택
y = data['Rings']

X = data.drop('Rings', axis = 1)
X[['M', 'F', 'I']] = X[['M', 'F', 'I']].astype(int)
```

```python
X
```

 **3) train/test set 분리**

```python
# 필요한 모듈 불러오기
from sklearn.model_selection import train_test_split
```

```python
# train과 test set 분리 (train:test = 7:3 비율로)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

**4) 선형회귀모델 생성, 모델 예측치 구하기**

```python
#필요한 모듈 불러오기
from sklearn.linear_model import LinearRegression
```

```python
#선형회귀모델 생성 및 훈련
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 모델 예측치 구하기
lr_pred = lr.predict(X_test)
# 모델 예측치를 활용해 최종적으로 전복의 나이를 예측
lr_pred + 1.5
```

**5) 모델 평가: MSE, RMSE, R2 score, corr 구하기**

```python
#필요한 모듈 불러오기
from sklearn.metrics import mean_squared_error, r2_score
```

- MSE, RMSE

```python
#mse, rmse
lr_mse = mean_squared_error(y_test, lr_pred)
lr_rmse = np.sqrt(lr_mse)
print(lr_mse, lr_rmse)
```

- R2 score

```python
#R2 score 측정
lr_r2 = r2_score(y_test, lr_pred)
print(lr_r2)
```

- 회귀 절편값

```python
#회귀 절편 값
intercept = lr.intercept_
print(intercept)
```

- 회귀 계수 값

```python
#회귀 계수 값
coef = lr.coef_
print(coef)
```

- 상관계수

Hint: corr 함수 이용.

```python
# 상관계수 구하기
result = pd.DataFrame({'y_test':y_test, 'pred':lr_pred})
result.corr()
```

# **2. Polynomial features**

```python
# PolynomialFeatures 라이브러리 호출
from sklearn.preprocessing import PolynomialFeatures
```

```python
# 임의 데이터 생성

X = np.arange(6).reshape(3, 2)

df =  pd.DataFrame(X)
df.columns = ['x_1','x_2']
df
```

```python
# 차원은 2로 설정
# fit_transform 메소드를 통해 데이터 변환
# PolynomialFeatures로 변환 된 데이터를 데이터 프레임 형태로 변환
poly = PolynomialFeatures(degree = 2)
poly.fit(X)
poly_ftr = poly.transform(X)
poly_ftr
```

```python
poly_ftr = pd.DataFrame(poly_ftr)
poly_ftr
```

```python
# df_poly의 컬럼을 1,x1,x2,x1^2,x1*x2,x2^2 로 변경
poly_ftr.columns = ['1','x1','x2','x1^2','x1*x2','x2^2']
poly_ftr
```
