## [EDA]
+ title : 영화의 제목
+ distributor : 배급사
+ genre : 장르
+ release_time : 개봉일
+ time : 상영시간(분)
+ screening_rat : 상영등급
+ director : 감독이름
+ dir_prev_bfnum : 해당 감독이 이 영화를 만들기 전 제작에 참여한 영화에서의 평균 관객수(단 관객수가 알려지지 않은 영화 제외)
+ dir_prev_num : 해당 감독이 이 영화를 만들기 전 제작에 참여한 영화의 개수(단 관객수가 알려지지 않은 영화 제외)
+ num_staff : 스텝수
+ num_actor : 주연배우수
+ box_off_num : 관객수

dir_prev_bfnum : 해당 감독이 이 영화를 만들기 전 제작에 참여한 영화에서의 평균 관객수(단 관객수가 알려지지 않은 영화 제외) -> train과 test 모두 dir_prev_bfnum이 결측값으로 존재하는 데이터들은 전부 dir_prev_num이 0인 영화들이다. 즉, 감독의 이전작품이 존재 하지 않은 경우이다. 따라서 dir_prev_bfnum에 있는 결측값은 전부 0으로 채운다.-> 이전에 제작에 참여한 영화가 없다고 하고 0으로 바꿈

```python
train['dir_prev_bfnum'].fillna(0, inplace=True)
```

```python
train['release_time']=pd.to_datetime(train['release_time'])
```

```python
train.plot(kind = 'scatter', x = 'num_staff', y = 'box_off_num')
plt.title('스태프 수와 관객 수 사이의 산점도')
```

```python
train[['release_time','box_off_num']].sort_values('release_time', ascending=False)
```

```python
train[['genre','box_off_num']].groupby('genre').mean().sort_values('box_off_num', ascending=False).plot(kind='bar')
plt.title('장르에 따른 평균 관객 수')
```

```python
train[['genre','num_staff']].groupby('genre').mean().sort_values('num_staff', ascending=False).plot(kind='bar')
plt.title('장르에 따른 평균 스태프 수')
```

```python
train[['director','box_off_num']].groupby('director').mean().sort_values('box_off_num', ascending=False)[:15].plot(kind='bar')
plt.title('감독에 따른 평균 관객 수 상위 15명')
```

```python
train[train['genre']==genre_list[0]]
```

```python
df1 = train[train['genre']==genre_list[0]].sort_values(by=['release_time'], axis=0)
df2 = train[train['genre']==genre_list[1]].sort_values(by=['release_time'], axis=0)
df3 = train[train['genre']==genre_list[2]].sort_values(by=['release_time'], axis=0)
```

```python
plt.plot(df1['release_time'], df1['box_off_num'], marker = 'o', markersize = 3)
plt.plot(df2['release_time'], df2['box_off_num'], marker = 'o', markersize = 3)
plt.plot(df3['release_time'], df3['box_off_num'], marker = 'o', markersize = 3)
plt.legend(labels = ['드라마', '다큐멘터리', '멜로/로맨스'], loc='best')  #범례 표시
plt.title('데이터 상 가장 많은 장르 3개의 개봉 시기에 따른 관객 수')

plt.show()
```

```python
df1 = train[train['genre']=='느와르'].sort_values(by=['release_time'], axis=0)
df2 = train[train['genre']=='액션'].sort_values(by=['release_time'], axis=0)
df3 = train[train['genre']=='SF'].sort_values(by=['release_time'], axis=0)
```

```python
plt.plot(df1['release_time'], df1['box_off_num'], marker = 'o', markersize = 3)
plt.plot(df2['release_time'], df2['box_off_num'], marker = 'o', markersize = 3)
plt.plot(df3['release_time'], df3['box_off_num'], marker = 'o', markersize = 3)
plt.legend(labels = ['느와르', '액션', 'SF'], loc='best')  #범례 표시
plt.title('평균 관객 수가 가장 많은 장르 3개의 개봉 시기에 따른 관객 수')


plt.show()
```
